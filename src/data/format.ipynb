```python
import pandas as pd

# Load your dataset
df = pd.read_csv('data.csv')  # Make sure 'data.csv' has the expected columns

# Step 1: Create a list of unique node names and map them to indexes
unique_nodes = sorted(set(df['Platform']).union(df['Brand']))
node_index = {name: i for i, name in enumerate(unique_nodes)}

# Step 2: Build the nodes list
nodes = [{"name": name} for name in unique_nodes]

# Step 3: Build the links list, using index references for source and target
links = [
    {
        "source": node_index[row["Platform"]],
        "target": node_index[row["Brand"]],
        "value": int(row["Brands included"]),
        "dealType": row["Deal type"]
    }
    for _, row in df.iterrows()
]

# Step 4: Combine into final Sankey-friendly structure
sankey_data = {
    "nodes": nodes,
    "links": links
}
```

```python
import pandas as pd
import json
from urllib.parse import urlparse, unquote

def extract_domain(url):
    try:
        if not url or not url.startswith("http"):
            return None

        parsed = urlparse(url)
        hostname = parsed.hostname

        # Handle Wayback Machine URLs
        if hostname == "web.archive.org":
            parts = parsed.path.split("/")[3:]
            original_url = "/".join(parts)
            if not original_url.startswith("http"):
                original_url = f"https://{original_url.lstrip('/')}"
            hostname = urlparse(unquote(original_url)).hostname

        return hostname.replace("www.", "") if hostname else None
    except Exception:
        return None

def extract_text_from_url(url):
    return extract_domain(url) or "link"

def transform_dataframe(df):
    result = []
    for _, row in df.iterrows():
        links = []

        # Prefer 'source_links' if it exists and is a list
        if isinstance(row.get("source_links"), list):
            urls = row["source_links"]
        elif pd.notna(row.get("source_url_1")):
            urls = [row["source_url_1"]]
            if pd.notna(row.get("source_url_2")):
                urls.append(row["source_url_2"])
        else:
            urls = []

        for url in urls:
            if pd.notna(url):
                links.append({
                    "text": extract_text_from_url(url),
                    "url": url
                })

        entry = {
            "date": row["date"],
            "platform": row["platform"] if isinstance(row["platform"], list) else [row["platform"]],
            "category": row["category"] if isinstance(row["category"], list) else [row["category"]],
            "title": row.get("AI incident title", "(Untitled)") or "(Untitled)",
            "description": row.get("desc", ""),
            "links": links,
            "last_updated": row["last_updated"]
        }
        result.append(entry)
    return result

# Example usage
# df = pd.read_csv("your_data.csv")  # Load your DataFrame
json_output = transform_dataframe(df[['date',	'platform', 'category', 'desc', 'source_links', 'last_updated']].replace({pd.NA: None, pd.NaT: None, float('nan'): None}))
# json_output
# To view:
print(json.dumps(json_output, indent=2))

# Save to JSON file
with open("data-timeline.json", "w") as f:
    json.dump(json_output, f, indent=2)
```
